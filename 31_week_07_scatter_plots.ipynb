## Continuous to Continuous relationships

The fundamental plot for showing the relationship between 2 continuous variables is the **SCATTER PLOT**.

In Seaborn, the FIGURE LEVEL function for showing RELAIONSHIPS is the `sns.relplot()` function.

The `x` axis is associated with a continuous variable and the `y` axis is associated with a second continuous variable.

```python
sns.relplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm')

plt.show()
```

```python
sns.relplot(data = penguins, x='bill_depth_mm', y='flipper_length_mm')

plt.show()
```

```python
penguins.info()
```

```python
sns.relplot(data = penguins, x='body_mass_g', y='bill_length_mm')

plt.show()
```

The underlying AXIS level function called by the FIGURE LEVEL `sns.relplot()` is the `sns.scatterplot()` function.

```python
fig, ax = plt.subplots()

sns.scatterplot(data = penguins, x='flipper_length_mm', y='bill_length_mm', ax=ax)

plt.show()
```

In Summary, the SCATTER PLOT shows the RELATIONSHIP between 2 continuous variables.

However, SCATTER PLOTS can show MORE THAN just 2 variables!!!!!

A very common practice is to INCLUDE a categorical variable via color!!! We can associate a categorical variable with the `hue` argument!!!!

Let's return to our first plot.

```python
sns.relplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm')

plt.show()
```

But we have categorical variables, object data types, in the `penguins` data set!

```python
penguins.info()
```

What if we GROUP BY a categorical variable? In other words...we can study if the RELATIONSHIP between the 2 variables CHANGES across CATEGORIES!!!!

Let's assign the `species` variable to the `hue` argument in `sns.relplot()`.

```python
sns.relplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species')

plt.show()
```

Instead of coloring by the categorical variable we can also use FACETS to SEPARATE the groups!

```python
sns.relplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm', col='species')

plt.show()
```

You can combine `hue` with `col` facets to get facets and colors.

```python
sns.relplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species', col='species')

plt.show()
```

However...if you do NOT color by the same variable as the FACET...that means you can use DIFFERENT categorical variables for color and facet!!!

You can GROUP the relationship on 2 separate and distinct categorical variables! You are grouping by the COMBINATION of 2 categorical variables!

The combination on `species` and `island` on the relationship!

```python
sns.relplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species', col='island')

plt.show()
```

We can add yet another categorical variable throw ROW facets!!!!

```python
penguins.info()
```

If we use ROW and COL facets and associate `hue` with a categorical variable we are effectively GROUPING the relationship between 2 continuous variables on the combination of 3 categorical variables!!!

```python
sns.relplot( data = penguins, x='flipper_length_mm', y='bill_depth_mm',
             hue='sex',
             col='island',
             row='species')

plt.show()
```

Remember to always check the COMBINATIONS between categoricals BEFORE looking at the relationships between continuous variables!!

```python
sns.catplot(data = penguins, x='island', hue='species', kind='count')

plt.show()
```
