## For-loop

```python
for n in range(7):
    # inside the body of the for-loop
    # actions occur for the CURRENT value of the iterating variable
    print( n )
```

```python
%whos
```

```python
for n in range(7):
    # inside the body of the for-loop
    # still inside the body
    print( n )
    # inside the body of the for loop
    print( n )
# we are OUTSIDE the body of the for-loop
# because we are NOT indented
# that means we are NO longer iterating
print( n )
print( n * 1000 )
# still outside the for loop!!
```

```python
%whos
```

```python
for n2 in range(7):
    # inside the for loop!
### OUTSIDE the for loop
print( n2 )
```

```python
for n2 in range(7):
    # inside the for loop!!
    print( 3 )
print( n2 )
```

```python
for n3 in range(7):
    # the for loops expects some kind of action!!
    n3 + 1
    # we simply adding a 1 to the iterating variable but NOT saving, defining, or printing it
### outside the for loop
print( n3 )
```

```python
%whos
```

```python
for nn in range(7):
    print( "Iterating variable value is: %d" % (nn) )
```

```python
%whos
```

```python
for mm in range(7):
    print( "Old iterating variable, nn, is: %d WHILE current iterating variable, mm, is: %d" % (nn, mm) )
```

```python
for zz in range(7):
    print( "%d + %d + %d = %d" % (nn, mm, zz, nn+mm+zz) )
```

## Nested for-loops

We can iterate over 2 sequences at the same time! Nested for-loops consist of OUTER and INNER for-loops!!!

```python
for o1 in range(3):
    # this is the OUTER for loop
    print( "Outer iteration number: %d" % (o1) )
    print( " -- ## -- " )
    for i1 in range(5):
        # this is inside the INNER for loop
        # it looks like a further SUB SECTION because we indented more!!!
        print( "Inner iteration number: %d" % (i1) )
        # still in the INNER for loop
    # we are BACK within the OUTER for loop!!!!
    print( " -- $$ -- " )
```

```python
for o2 in range(5):
    # outer for loop iterates over an INTEGER
    print("Outer iteration number: %d" % (o2) )
    for i2 in ['a', 'b', 'c', 'd']:
        # INNER for loop
        print( o2 * i2 )
    # back in the OUTER for loop
    print( ' -- $$ -- ' )
```

```python
for o3 in ['a', 'b', 'c', 'd']:
    for i3 in ['AA', 'BB', 'CC', 'DD']:
        print( (o3, i3, o3 + ' ' + i3 + ' ' + o3) )
```

## List comprehension

The list comprehension is a streamlined way of managing iterating.

List comprehensions require that we completely understand the 3 basic elements of iteration:

* the action to repeat
* the iterating variable that changes
* the sequence to iterate over

In code, the syntax is:

`<output list> = [ <action> for <variable> in <sequence> ]`

What is the ACTION we are repeating???  We are calling `random.random()` many times!

What is the iterating variable? We were calling it `n` but it doesn't really do anything here...

What is our sequence to iterate over? The sequence is generated by `range()`!!!

```python
random.seed(2100)

y = [ random.random() for iterating_variable in range(11) ]
```

```python
type( y )
```

```python
len( y )
```

```python
y
```

```python
y == xu_c
```

The iterating variable is not really important in our current application. More often than not, the iterating variable IS important. But when it is not needed, a common practice is to use the UNDERSCORE, `_`, as the iterating variable.

```python
random.seed(2100)

y2 = [ random.random() for _ in range(11) ]
```
