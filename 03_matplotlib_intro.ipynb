## Overview

matplotlib has 2 main styles of programming.

The first or classic is the STATEFUL style which is based on MATLAB.

The second approach is the OBJECT ORIENTED STYLE approach.

## Stateful or MATLAB style

* step 1: initiate the figure
* step 2: specify the subplot that contains the data
* step 3: populate the data within the plot (plot the data)
* step 4: show the figure

Let's begin with step 1 and work through each step one at a time.

```python
### step 1 - initiate the figure

plt.figure()
```

Why am I making the big deal about the FIGURE vs SUBPLOT (AXES)????

Because we can have multiple subplots associated with a single figure!!!!

Let's initiate a single figure that has 2 subplots. One shows the SINE WAVE and another shows a COSINE WAVE.

```python
plt.figure()

### define 2 subplots contained in this figure
plt.subplot( 2, 1, 1 )

plt.plot( np.sin( np.linspace(0, 10, 100) ) )

### define the SECOND subplot within the figure
plt.subplot( 2, 1, 2 )

plt.plot( np.cos( np.linspace(0, 10, 100) ) )

plt.show()
```

Examine this again but only populate data on the 2nd subplot.

```python
plt.figure()

plt.subplot(2, 1, 1)

plt.subplot(2, 1, 2)

plt.plot( np.cos( np.linspace(0, 10, 100) ) )

plt.show()
```

We could instead use 1 row and 2 columns of subplots.

```python
plt.figure()

plt.subplot(1, 2, 1)

plt.plot( np.sin( np.linspace(0, 10, 100) ) )

plt.subplot(1, 2, 2)

plt.plot( np.cos( np.linspace(0, 10, 100) ) )

plt.show()
```

Highlight we can control the exact AXES or subplot the data are populated to!

```python
plt.figure()

plt.subplot(1, 5, 1)

plt.plot( np.sin( np.linspace(0, 10, 100) ) )

plt.subplot(1, 5, 2)

plt.plot( np.cos( np.linspace(0, 10, 100) ) )

plt.show()
```

```python
plt.figure()

plt.subplot(1, 5, 1)

plt.plot( np.sin( np.linspace(0, 10, 100) ) )

plt.subplot(1, 5, 5)

plt.plot( np.cos( np.linspace(0, 10, 100) ) )

plt.show()
```

We can control the figure size with the `figsize` argument to the `plt.figure()` function!

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 5, 1)

plt.plot( np.sin( np.linspace(0, 10, 100) ) )

plt.subplot(1, 5, 5)

plt.plot( np.cos( np.linspace(0, 10, 100) ) )

plt.show()
```

## Object Oriented Style

Many aspects of the Object Oriented Style figures are the same as the STATEFUL figures. But the way we access or apply functions to them are different!

* step 1: Define the figure and axes objects - create variables!
* step 2: plot the data within the axes using methods associated with the axes objects!
* step 3: modify teh axes object attributes if needed
* step 4: modify the figure object attributes if needed
* step 5: show the figure

The motivation for the object oriented approach is that the STATEFUL approach works with HIDDEN OBJECTS!!!!

Define variables or objects for the FIGURE and AXES that we are modifying!

The common names for these objects are `fig` and `ax`

```python
%whos
```

```python
fig, ax = plt.subplots()
```

We have a lot of control over the subplots or axes within a figure.

Let's make a figure with 2 rows and 3 columns of subplots.

```python
fig, axs = plt.subplots(2, 3)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(12, 6))
```

We can force each subplot to SHARE axes!!!!

```python
fig, axs = plt.subplots(2, 3, figsize=(12, 6), sharex=True, sharey=True )
```

```python
type( axs )
```

```python
axs.shape
```

We can slice the 2D numpy array just like we slice any 2D numpy array!

```python
axs[0, 0]
```

```python
axs[0, 1]
```

```python
print( type( axs[0, 1] ) )
```

Use a nested for loop to iterate over each subplot within the figure and display some text.

```python
for i in range( axs.shape[0] ):
    for j in range( axs.shape[1] ):
        axs[i, j].text( 0.5, 0.5, str( (i,j) ), ha='center', fontsize=20)
```

```python
fig
```

Put everything in a single cell.

```python
fig, axs = plt.subplots(2, 3, figsize=(12, 4), sharex=True, sharey=True)

for i in range( axs.shape[0] ):
    for j in range( axs.shape[1] ):
        axs[ i, j ].text( 0.5, 0.5, str( (i, j) ), ha='center', fontsize=20)

plt.show()
```
