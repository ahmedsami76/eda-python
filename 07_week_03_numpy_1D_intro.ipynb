The `np.arange()` function has 3 arguments.

START, STOP (EXCLUSIVE), and STEP.

```python
np.arange(6)
```

```python
np.arange(0, 6, 3)
```

```python
np.arange(0, 6, 2)
```

```python
np.arange(0, 100, 20)
```

```python
np.arange(0, 100+1, 20)
```

The `np.linspace()` function is very similar to `np.arange()` BUT...

START, STOP (INCLUSIVE), and Number of points.

```python
np.linspace(0, 6, num=1)
```

```python
np.linspace(0, 6, num=6)
```

```python
np.linspace(0, 6, num=6+1)
```

```python
np.linspace(0, 6, num=20)
```

## Differences with lists

Operators may behave differently on NumPy arrays compared to lists.

```python
a_list
```

```python
a_list == [0, 1, 2, 3, 4, 5]
```

```python
a
```

```python
a == np.array([0, 1, 2, 3, 4, 5])
```

```python
a == np.array([-10, 1, 2, 3, 4, 5])
```

```python
np.zeros(5) == np.zeros(10)
```

```python
np.array_equal( a, np.array([0, 1, 2, 3, 4, 5]))
```

```python
np.array_equal( a, np.array([-10, 1, 2, 3, 4, 5]))
```

The `+` operator behaves differently!!!

```python
[1, 2, 3] + [3, 2, 1]
```

```python
np.array([1, 2, 3]) + np.array([3, 2, 1])
```

```python
np.zeros(5) + np.ones(5)
```

```python
np.zeros(10) + np.ones(5)
```
