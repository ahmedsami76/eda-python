### Visualize continuous to continuous relationships via TREND PLOTS

The TREND PLOT is an extension of the scatter plot that adds further "quantitative" information about the relationship between two continuous variables.

## Import Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

## Read data

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.info()
```

## Scatter plot

Show the relationship between two continuous variables.

```python
sns.relplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm')

plt.show()
```

We saw that by GROUPING or CONDITIONING on a categorical variable...our interpetation of the scatter plot changed!

```python
sns.relplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species')

plt.show()
```

## Trend plots

We can add TREND LINES to visualize the **LINEAR** relationship between the two continuous variables.

TREND LINES correspond to the **BEST FIT LINE** between the two variables.

Later on...we will learn HOW this is line is created...but for now let's visualize it and let the line or TREND HELP US interpret the relationship!

The TREND is NOT created with the `sns.relplot()` function.

Instead, the TREND is created with the FIGURE LEVEL function `sns.lmplot()`.

```python
sns.lmplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm')

plt.show()
```

We are viewing an ANTI or NEGATIVE relationship between the two variables!!!!

BUT...what happens if we GROUP or CONDITION ON a categorical variable?!??!?

Let's COLOR via the `hue` argument BY `species`.

```python
sns.lmplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species')

plt.show()
```

GIVEN or GROUPING BY or CONDITIONING ON the `species` CHANGES our opinion about the RELATIONSHIP betwen the 2 continuous variables! GROUPING BY `species` reveals a POSITIVE relationship between the two!!!

We can of course create the TREND PLOT between 2 different continuous variabels!

```python
penguins.info()
```

```python
sns.lmplot(data = penguins, x='body_mass_g', y='bill_length_mm')

plt.show()
```

```python
sns.lmplot(data = penguins, x='body_mass_g', y='bill_length_mm', hue='species')

plt.show()
```

As with the scatter plot we can also GROUP BY or CONDITION ON a SECONDARY VARIABLE via FACETS!!!!!

Let's facet by the penguin `sex`.

```python
sns.lmplot(data = penguins, x='body_mass_g', y='bill_length_mm', hue='species', col='sex')

plt.show()
```

```python
sns.lmplot(data = penguins, x='body_mass_g', y='bill_length_mm', hue='sex', col='species')

plt.show()
```

A THIRD categorical variable can be included via ROW facets!

```python
sns.lmplot(data = penguins, x='body_mass_g', y='bill_length_mm', hue='sex', col='species', row='island')

plt.show()
```

```python
sns.lmplot(data = penguins, x='flipper_length_mm', y='bill_depth_mm', hue='sex', col='species', row='island')

plt.show()
```

The TREND PLOT does NOT connect the DOTS!!! The TREND PLOT is actual a MODEL and thus a MODEL is fit behind the scenes!
